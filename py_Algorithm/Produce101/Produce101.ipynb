```python
Data=[]
Tmp=openpyxl.load_workbook('Data.xlsx')
Excel=Tmp.active

for menu1 in Excel.rows:
    i=0
    temp=[]
    for menu0 in menu1:
        temp.append(menu1[i].value)
        i+=1
    Data.append(temp)

    
Data.pop(0)
print(Data)
```
