```python
def Checking(Num):
    global Collect
    global people
    global Pass
    
    for i in range(len(Pass)):
        if Pass[i]==0:
            for j in range(len(Collect[Num])):
                if people[i][Num]>people[Collect[Num][j]][Num]:
                    Pass[Collect[Num][j]]=0
                    Pass[i]=Num
                    if Collect[Num][j]==0:
                        Collect[Num][j]=i
                    else:
                        Collect[Num].insert(j,i)
                        Collect[Num].pop()
                        
                    break
        print(Collect)
    
    if Num<2: return
    
    for i in range(1,Num+1):
        for j in range(len(Collect[i])):
            for r in range(len(Collect[Num])):
                if people[Collect[i][j]][i]-people[Collect[Num][r]][i] <people[Collect[i][j]][Num]-people[Collect[Num][r]][Num]:
                    Pass[Collect[i][j]]=Num
                    Pass[Collect[Num][r]]=i
                    Collect[Num][r],Collect[i][j]=Collect[i][j],Collect[Num][r]
```

```python
N=int(input("Number of N: "))

Data=[]

for i in range(N):
    Tmp=[]
    Tmp.append(int(input("Score of Document: ")))
    Tmp.append(int(input("Score of Personality: ")))
    Tmp.append(int(input("Score of Aptitude: ")))
    Tmp.append(int(input("Score of Interview: ")))
    Tmp.append(int(input("Score of Coding: ")))
    
    Data.append(Tmp)
```

```python
Pass=[0 for i in range(N+1)]

Collect=[]
Collect.append([])
Collect.append([0 for i in range(int(input("Typing Num of HR TO: ")))])
Collect.append([0 for i in range(int(input("Typing Num of Development TO: ")))])
Collect.append([0 for i in range(int(input("Typing Num of Strategic Planning TO: ")))])
Collect.append([0 for i in range(int(input("Typing Num of Finance TO: ")))])

people=[]
people.append([-1 for i in range(5)])
for i in Data:
    Tmp=[]
    Tmp.append([])
    Tmp.append(i[1])
    Tmp.append(i[4])
    Tmp.append(int((i[0]*0.15+i[1]*0.15+i[2]*0.2+i[3]*0.1+i[4]*0.4)*100)/100)
    Tmp.append(int((i[0]*0.2+i[1]*0.2+i[2]*0.4+i[3]*0.2)*100)/100)
    people.append(Tmp)
```

```python
for i in range(1,5): 
    Checking(i)
    print("#########%d#########"%i)
    
print(Collect)
```

```python
w=0
r=-1
for i in Collect:
    r+=1
    for j in i:
        w+=people[j][r]

print("Value of w: %.02f"%w)
```
