```python
import sys

sys.setrecursionlimit(100000000)
```

```python
def sol(index, now, visited):
    if now<0 or now>9: return 0
    elif index==n:
        if(((1<<10)-1)==visited): return 1
        
        
    ret=dp[index][now][visited]
    if ret !=-1: return ret
    if now==0: 
        dp[index][now][visited]=sol(index+1,now+1,visited |(1<<(now+1))) + sol(index+1,now-1,visited) % 1000000000
    else:
        dp[index][now][visited]=sol(index+1,now+1,visited |(1<<(now+1))) + sol(index+1,now-1,visited|(1<<(now-1))) % 1000000000
    return dp[index][now][visited]%1000000000
```

```python
n=input()

ans=0

dp=[[[-1 for i in range(2**11)] for j in range(10)] for k in range(102)]

for i in range(1,10):
    ans+=sol(1,i,1<<i)
    ans %= 1000000000
print(ans)
```
