데이터 읽기

```python
Data=[]
Tmp=openpyxl.load_workbook('Data.xlsx')
Excel=Tmp.active

for menu1 in Excel.rows:
    i=0
    temp=[]
    for menu0 in menu1:
        temp.append(menu1[i].value)
        i+=1
    Data.append(temp)

    
Data.pop(0)
print(Data)
```

```python
Answer=[]

for studentB in Data:
    studentA=[]
    studentA.append(studentB[0])
    score=0
    while True:
        if studentB[1]<11:
            studentA.append(-20000)
            break
        else:
            score+=(14-studentB[1])*(-5)
        if studentB[2]<20:
            studentA.append(-20000)
            break
        else:
            score+=(22-studentB[2])*(-5)
        if studentB[3]<60:
            studentA.append(-20000)
            break
        else:
            score+=studentB[3]
        if studentB[4]<80:
            studentA.append(-20000)
            break
        else:
            score+=studentB[4]
        if studentB[5]<60:
            studentA.append(-20000)
            break
        else:
            score+=studentB[5]
            studentA.append(score)
            break
    Answer.append(studentA)

print(Answer)
```

```python
ANSWER=sorted(Answer,reverse=True, key=lambda data:data[1])

for i in ANSWER:
    print(i)
    
i=0
for student in ANSWER:
    if i<len(ANSWER)/2:
        ANSWER[i][1]="A+"
    elif ANSWER[i][1] == -20000:
        ANSWER[i][1]="F"
    elif i<len(ANSWER)*0.9:
        ANSWER[i][1]="B+"
    else:
        ANSWER[i][1]="C+"
    i+=1
    
for i in ANSWER:
    print(i)
```
